```python
import pandas as pd
import numpy as np

# Sample DataFrame with outliers
data = {'A': [1, 2, 3, 4, 5, 100]}
df = pd.DataFrame(data)

# Calculate Z-score for each data point in column 'A'
z_scores = np.abs((df['A'] - df['A'].mean()) / df['A'].std())

# Define threshold for outlier detection (e.g., Z-score > 3)
threshold = 3
outliers = df[z_scores > threshold]

print("Outliers:")
print(outliers)
```
