```python
import pandas as pd
import numpy as np

# URL of the dataset
url = "https://storage.googleapis.com/download.tensorflow.org/data/iris_training.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(url)

# Display the original DataFrame
print("Original DataFrame:")
print(df.head())  # Display the first few rows
print()

# Adding a new column
new_col = np.random.randint(100, size=len(df))
df.insert(3, 'new_col2', new_col)  # Example values

# Display the DataFrame after adding the new column
print("DataFrame after adding 'New_Column':")
print(df.head())  # Display the first few rows
print()

# Accessing rows
print("Accessing rows:")
print(df.iloc[0])  # Access the first row
print(df.iloc[1:3])  # Access rows from index 1 to 2
print()

# Removing a column
df.drop(columns=['setosa'], inplace=True)  # Specify the column to delete

# Display the DataFrame after removing the specified column
print("DataFrame after removing 'Column_To_Delete':")
print(df.head())  # Display the first few rows
```
