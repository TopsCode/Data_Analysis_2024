```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(1)
data = np.random.randn(100)

# Create a violin plot
plt.figure()
plt.violinplot(data, showmedians=True)

# Set plot labels and title
plt.xlabel('Variable')
plt.ylabel('Value')
plt.title('Univariate Violin Plot')

# Show the plot
plt.show()
```

Violin Plot:
A violin plot is similar to a box plot but provides a deeper insight into the data distribution by displaying a kernel density estimation (KDE) of the underlying distribution. It combines a box plot with a rotated kernel density plot on each side.

Box Plot:
A box plot displays the distribution of data based on a five-number summary: minimum, first quartile (Q1), median (second quartile or Q2), third quartile (Q3), and maximum. It helps identify outliers and provides a visual representation of the data's spread and central tendency.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Generate some example data
data = [82, 85, 88, 90, 90, 92, 93, 94, 95, 100, 102]

# Create a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=data)
plt.title('Box Plot')
plt.xlabel('Data')
plt.show()
```
